Resumeix gràficament el data set DelayedFlights.csv

Crea almenys una visualització per:

- Una variable categòrica (UniqueCarrier)
- Una variable numèrica (ArrDelay)
- Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
- Dues variables numèriques (ArrDelay i DepDelay)
- Tres variables (ArrDelay, DepDelay i UniqueCarrier)
- Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#data set net després de l'exploració tasca 5
delay_df = pd.read_csv('delay_df.csv')
```

```python
delay_df.head()
```

```python
#Una variable categòrica (UniqueCarrier)

plot1 = sns.countplot(x='UniqueCarrier', data=delay_df) #nº de vuelos / compañía
fig1 = plot1.get_figure()
fig1.savefig("plot1.png")
```

```python
plt.figure(figsize=(6,6))
plt.pie(delay_df.UniqueCarrier.value_counts(), labels= delay_df.UniqueCarrier.unique()) #nº vuelos/compañía

plt.savefig("plot2.png")
```

```python
#Una variable numèrica (ArrDelay)

plt.figure(figsize=(8,6))
sns.displot(data=delay_df, x="ArrDelay") #densidad de valores
plt.savefig("plot3.png")
```

```python
#observamos skew a la izquierda
plt.figure(figsize=(8,6))
sns.kdeplot(delay_df['ArrDelay'])
plt.savefig("plot4.png")
```

```python
delay_df.describe().round(2)
```

```python
#observar la densidad de ArrDelay entre el primer y tercer cuartil
arr_delay = delay_df[(delay_df['ArrDelay'] <= delay_df['ArrDelay'].quantile(0.75)) & (delay_df['ArrDelay'] >= delay_df['ArrDelay'].quantile(0.25))]
plot5 = sns.kdeplot(arr_delay['ArrDelay'])
fig5 = plot5.get_figure()
fig5.savefig("plot5.png")
```

```python
#Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)

#esta gráfica nos muestra la media de retraso de cada compañía aérea

plt.figure(figsize=(8,6))
plot6 = sns.barplot(x='UniqueCarrier', y='ArrDelay', data=delay_df)
plt.savefig("plot6.png")
```

```python
#nos muestra cómo de regular es el retraso de cada compañía basándose en la desviación estándar

plt.figure(figsize=(8,8))
plot7 = sns.barplot(x='UniqueCarrier', y='ArrDelay', data=delay_df, estimator=np.std) 
plt.savefig("plot7.png")
```

```python
#Dues variables numèriques (ArrDelay i DepDelay)

plt.figure(figsize=(8,6))
plot8 = sns.relplot(x='ArrDelay', y='DepDelay', data=delay_df)
plt.savefig("plot8.png")
```

```python
#Tres variables (ArrDelay, DepDelay i UniqueCarrier)

sns.set_context('paper', font_scale=1.4)
plot9 = sns.relplot(x='ArrDelay', y='DepDelay', data=delay_df, hue='UniqueCarrier')
plt.savefig("plot9.png")
```

```python
#Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).

plot10 = sns.pairplot(delay_df, vars=['ArrDelay', 'DepDelay', 'AirTime'], hue='UniqueCarrier')
plt.savefig("plot10.png")
```
